```
class GroceryItem:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def __del__(self):
        pass
    def calculate_total(self):
        return self.price * self.quantity

    def __str__(self):
        return f"{self.name.ljust(10)} | PHP {str(self.price).ljust(5)} | {str(self.quantity).ljust(8)} | PHP {self.calculate_total():.2f}"

    def __copy__(self):
        return type(self)(self.name, self.price, self.quantity)

    def copy_assign(self, other):
        self.name = other.name
        self.price = other.price
        self.quantity = other.quantity

class Fruit(GroceryItem):
    def __str__(self):
        return super().__str__() + " | Fruit"

class Vegetable(GroceryItem):
    def __str__(self):
        return super().__str__() + " | Vegetable"

def display_grocery_list(items):
    print("\n" + "="*60)
    print("JENNA'S GROCERY LIST".center(60))
    print("="*60)
    print("Item       | Price  | Quantity | Subtotal  | Type")
    print("-"*60)
    for item in items:
        print(item)
    print("-"*60)

def calculate_grand_total(items):
    return sum(item.calculate_total() for item in items)

def main():
    # Problem 1: Class implementation
    # Problem 2: Create and display grocery list
    grocery_list = [
        Fruit("Apple", 10, 7),
        Fruit("Banana", 10, 8),
        Vegetable("Broccoli", 60, 12),
        Vegetable("Lettuce", 50, 10)
    ]

    display_grocery_list(grocery_list)

    # Problem 3: Calculate total sum
    total = calculate_grand_total(grocery_list)
    print(f"GRAND TOTAL: ₱{total:.2f}".center(60))
    print("="*60 + "\n")

    # Problem 4: Remove Lettuce
    print("Removing Lettuce from the list...\n")
    grocery_list = [item for item in grocery_list if item.name != "Lettuce"]

    display_grocery_list(grocery_list)
    new_total = calculate_grand_total(grocery_list)
    print(f"UPDATED GRAND TOTAL: ₱{new_total:.2f}".center(60))
    print("="*60)

if __name__ == "__main__":
    main()
```

```
class GroceryItem:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def calculate_sum(self):
        return self.price * self.quantity

    def __str__(self):
        return f"{self.name} - Price: {self.price}, Quantity: {self.quantity}, Total: {self.calculate_sum()}"

class Fruit(GroceryItem):
    def __init__(self, name, price, quantity):
        super().__init__(name, price, quantity)

class Vegetable(GroceryItem):
    def __init__(self, name, price, quantity):
        super().__init__(name, price, quantity)


# Original grocery list
grocery_list = [
    Fruit("Apple", 10, 7),
    Fruit("Banana", 10, 8),
    Vegetable("Carrot", 60, 12),
    Vegetable("Lettuce", 50, 10),
]

# Function to calculate total sum
def total_sum(groceries):
    return sum(item.calculate_sum() for item in groceries)

# Visualization function
import matplotlib.pyplot as plt

def plot_grocery_costs(groceries, title):
    names = [item.name for item in groceries]
    costs = [item.calculate_sum() for item in groceries]
    plt.figure(figsize=(6,4))
    plt.bar(names, costs)
    plt.title(title)
    plt.xlabel("Item")
    plt.ylabel("Total Cost")
    plt.show()

# Plot before deletion
plot_grocery_costs(grocery_list, "Grocery Costs Before Deletion")
total_before = total_sum(grocery_list)

# Delete Lettuce
grocery_list = [item for item in grocery_list if item.name.lower() != "lettuce"]

# Plot after deletion
plot_grocery_costs(grocery_list, "Grocery Costs After Deletion")
total_after = total_sum(grocery_list)

total_before, total_after
```
